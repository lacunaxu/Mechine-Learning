By researching from online

Exact Match is a strict evaluation indicator for multi-label classification, requiring all labels to be completely correct for classification to be correct.
For multi-label data, Exact Match is equal to the proportion of labels that are completely matched in all samples.

Exact Match = Number of samples with all labels correctly predicted / Total number of samples

Hamming Score / Hamming Loss

Hamming Loss is a loose evaluation indicator in multi-label classification, which indicates the proportion of incorrectly predicted labels to the total number of labels.

1/N*L * sum(i=1 to N) sum(j=1 to L) l(yij != y_hatij)

Hamming Score is a supplement to Hamming Loss, which is used to measure the proportion of correctly predicted labels to the total number of labels.

Hamming Score = 1 − Hamming Loss

```python
def train_svm(X_train, y_train, X_test, y_test, class_labels, use_standardized=False, penalized=False, use_SMOTE=False):
    
    model_name = "LinearSVC with L1 penalty" if penalized else "SVC with RBF kernel"
    method_info = f"Standardized: {use_standardized}, SMOTE: {use_SMOTE}"
    print(f"Training {model_name} ({method_info}) for {y_train.name}")
    
    if use_standardized:
        scaler = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_test = scaler.transform(X_test)

    if use_SMOTE:
        smote = SMOTE(random_state=14)
        X_train, y_train = smote.fit_resample(X_train, y_train)

    if penalized:
        model = LinearSVC(penalty='l1', dual=False, max_iter=50000)
        param_grid = {'C': [0.01, 0.1, 1, 10, 100]}
    else:
        model = SVC(kernel='rbf', probability=True)
        param_grid = {
            'C': [0.01, 0.1, 1, 10, 100],
            'gamma': [0.001, 0.01, 0.1, 1, 10]
        }

    grid_search = GridSearchCV(model, param_grid, cv=10, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)

    print(f"Best Parameters for {y_train.name}: {grid_search.best_params_}")
    
    best_model = grid_search.best_estimator_
    y_test_pred = best_model.predict(X_test)
    y_test_prob = best_model.predict_proba(X_test) if hasattr(best_model, "predict_proba") else None

    plot_graph(y_test, y_test_pred, y_test_prob, class_labels) 

    return best_model
```

```python
def plot_graph(y_test, y_test_pred, y_test_prob, class_labels):
    Confusion_matrix = confusion_matrix(y_test, y_test_pred)
    print(f"\nConfusion Matrix is: {Confusion_matrix}")

    precision = precision_score(y_test, y_test_pred, average='weighted')
    recall = recall_score(y_test, y_test_pred, average='weighted')
    print(f"Overall Precision is : {precision:.4f}, Overall Recall is: {recall:.4f}")

    accuracy = accuracy_score(y_test, y_test_pred)
    print(f"Accuracy: {accuracy:.4f}")

    if y_test_prob is not None:
        for i, label in enumerate(class_labels):
            fpr, tpr, _ = roc_curve((y_test == label).astype(int), y_test_prob[:, i])
            auc = roc_auc_score((y_test == label).astype(int), y_test_prob[:, i])

            plt.figure()
            plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
            plt.plot([0, 1], [0, 1], linestyle='--', color='red')
            plt.xlabel("False Positive Rate")
            plt.ylabel("True Positive Rate")
            plt.title(f"ROC Curve for Class {label}")
            plt.legend(loc="lower right")
            plt.grid(True, linestyle='--', alpha=0.7)
            plt.show()
    else:
        print("Probability predictions not available for ROC/AUC.")
```

```python
#2(abc together)
def Monte_Carlo_Simulation(X, y):
    hamming_distances = []
    hamming_scores = []
    hamming_losses = []

    for _ in range(50):
        best_k = 0
        best_score = -1

        #2(a)
        for k in range(2, 51):
            labels = KMeans(n_clusters=k, random_state=12, n_init=10).fit_predict(X)
            score = silhouette_score(X, labels)
            if score > best_score:
                best_k = k
                best_score = score

        cluster_labels = KMeans(n_clusters=best_k, random_state=12, n_init=10).fit_predict(X)

        #2(b)
        majority_labels = {}
        for i in y.columns:
            majority_labels[i] = []
            for cluster in range(best_k):
                cluster_indices = np.where(cluster_labels == cluster)[0]
                if len(cluster_indices) > 0:
                    majority_label = y.iloc[cluster_indices][i].mode()[0]
                    majority_labels[i].append(majority_label)

        predicted_labels = pd.DataFrame(index=y.index, columns=y.columns)

        for i in range(len(cluster_labels)):
            cluster = cluster_labels[i]
            for col in y.columns:
                predicted_labels.at[i, col] = majority_labels[col][cluster]

        predicted_labels = predicted_labels.to_numpy()
        true_labels = y.to_numpy()

        #2(c)
        hamming_distance = np.mean(np.mean(true_labels != predicted_labels, axis=1))
        hamming_score = 1 - hamming_distance

        hamming_loss_value = np.mean([hamming_loss(true_labels[:, i], predicted_labels[:, i]) for i in range(true_labels.shape[1])])

        hamming_distances.append(hamming_distance)
        hamming_scores.append(hamming_score)
        hamming_losses.append(hamming_loss_value)
        
    print(f"Average Hamming Distance: {np.mean(hamming_distances):.4f} ± {np.std(hamming_distances):.4f}")
    print(f"Average Hamming Score: {np.mean(hamming_scores):.4f} ± {np.std(hamming_scores):.4f}")
    print(f"Average Hamming Loss: {np.mean(hamming_losses):.4f} ± {np.std(hamming_losses):.4f}")


Monte_Carlo_Simulation(X, y)
```
